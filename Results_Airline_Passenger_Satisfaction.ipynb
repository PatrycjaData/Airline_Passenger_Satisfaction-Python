```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('test.csv')
data.head(50)
```

```python
female_filter=data["Gender"].isin(["Female"])
female_filter
```

```python
female_data=data[female_filter].head(50)
female_data
```

```python
satisfaction_female=female_data.groupby('satisfaction').size()
satisfaction_female
```

```python
satisfaction_female.plot.pie(autopct ='%1.1f%%',
                            figsize = (6, 6),
                            explode = (0, 0.1),
                            cmap = 'Set3',
                            shadow =True,
                            startangle=90)
plt.title('Passenger Satisfaction - Female Group')
plt.show()                          
```

```python
male_filter=data["Gender"].isin(["Male"])
```

```python
male_data=data[male_filter].head(50)
```

```python
satisfaction_male=male_data.groupby('satisfaction').size()
satisfaction_male
```

```python
satisfaction_male.plot.pie(autopct='%1.1f%%',
                          figsize = (6, 6),
                          explode = (0, 0.1),
                          cmap = 'Set3',
                          shadow=True,
                          startangle=90)
plt.title('Passenger Satisfaction - Male Group')
plt.show()
```

```python
satisfaction_data=data.groupby(['Gender','satisfaction']).size()
satisfaction_data
```

```python
satisfaction = ["neutral or dissatisfied", "satisfied"]
female_counts = [7437, 5735]
male_counts = [7136, 5668]

x = np.arange(len(satisfaction))
width = 0.20  

fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, female_counts, width, label='female')
rects2 = ax.bar(x + width/2, male_counts, width, label='male')
ax.set_xlabel('satisfaction')
ax.set_ylabel('Number of passengers')
ax.set_title('Airline Passenger Satisfaction')
ax.set_xticks(x)
ax.set_xticklabels(satisfaction)
ax.legend()
ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)

plt.show()
```
